```python
import pandas as pd

def concatenate_row(row):
    # Concatenate relevant columns into a single string
    row_data = f"Product Name: {row['product_name']}\n" \
               f"Brand: {row['brand']}\n" \
               f"Category: {row['product_category_tree']}\n" \
               f"Retail Price: {row['retail_price']}\n" \
               f"Discounted Price: {row['discounted_price']}\n" \
               f"Rating: {row['product_rating']}\n" \
               f"Overall Rating: {row['overall_rating']}\n" \
               f"Description: {row['description']}\n"
    
    return row_data

# Apply the concatenation to each row
df['concatenated'] = df.apply(concatenate_row, axis=1)

# Collect all concatenated rows into a list
documents = df['concatenated'].tolist()
```

```python
import os
import streamlit as st
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
import google.generativeai as genai
from dotenv import load_dotenv
import pandas as pd
from langchain.schema import Document
from tqdm import tqdm

# Load environment variables
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

# Configure Google Generative AI
genai.configure(api_key=GOOGLE_API_KEY)

# Define the embedding model
embedding_model_name = "all-MiniLM-L6-v2"
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name)

# Create documents from CSV
def create_document(row):
    content = f"Product Name: {row['product_name']}\n" \
              f"Brand: {row['brand']}\n" \
              f"Category: {row['product_category_tree']}\n" \
              f"Retail Price: {row['retail_price']}\n" \
              f"Discounted Price: {row['discounted_price']}\n" \
              f"Rating: {row['product_rating']}\n" \
              f"Overall Rating: {row['overall_rating']}\n" \
              f"Description: {row['description']}"
    
    metadata = {
        "product_name": row['product_name'],
        "brand": row['brand'],
        "category": row['product_category_tree'],
        "retail_price": row['retail_price'],
        "discounted_price": row['discounted_price'],
        "product_rating": row['product_rating'],
        "overall_rating": row['overall_rating']
    }
    
    return Document(page_content=content, metadata=metadata)

# Create documents from CSV file
def load_documents():
    df = pd.read_csv('productData.csv')
    documents = []
    for _, row in tqdm(df.iterrows(), total=df.shape[0], desc="Creating Documents"):
        documents.append(create_document(row))
    return documents

# Load or create the vector store
@st.cache_resource
def load_vector_store():
    persist_directory = "chroma_db"
    if not os.path.exists(persist_directory):
        documents = load_documents()
        vector_store = Chroma.from_documents(
            documents=documents,
            embedding=embeddings,
            persist_directory=persist_directory
        )
        vector_store.persist()
    else:
        vector_store = Chroma(persist_directory=persist_directory, embedding_function=embeddings)
    return vector_store

# Load the vector store
vector_store = load_vector_store()

# Check the number of documents in the loaded vector store
st.write(f"Number of documents in the loaded vector store: {vector_store._collection.count()}")

# Create a conversation memory
memory = ConversationBufferMemory(k=5)  # Keeps the last 5 conversations

# Create a prompt template for query rephrasing
rephrase_template = """Given the following conversation history and the current query, rephrase the query to include relevant context from the conversation:

Conversation history:
{history}

Current query: {query}

Rephrased query:"""

rephrase_prompt = PromptTemplate(
    input_variables=["history", "query"],
    template=rephrase_template
)

# Create a prompt template for the RAG response
rag_template = """You are an AI assistant for a product recommendation system. Use the following pieces of context to answer the human's question. If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context: {context}

Human: {query}
Assistant:"""

rag_prompt = PromptTemplate(
    input_variables=["context", "query"],
    template=rag_template
)

# Streamlit app layout
st.title("Product Recommendation AI Assistant")

# Input from the user
query = st.text_input("Enter your query:")

if query:
    # Retrieve relevant context from the vector store using the embeddings
    docs = vector_store.similarity_search(query, k=5)  # Fetch top 5 relevant documents
    context = "\n".join([doc.page_content for doc in docs])

    # Display retrieved context
    st.write("### Retrieved Context")
    st.write(context)

    # Rephrase the query based on the conversation history
    history = "\n".join(memory.load_memory_variables({})["buffer"])  # Get conversation history
    rephrased_query = rephrase_prompt.format(history=history, query=query)

    # Call the Google Generative AI model with the rephrased query
    final_prompt = rag_prompt.format(context=context, query=rephrased_query)
    response = genai.GenerativeModel(model_name="gemini-1.5-flash").predict(final_prompt)

    # Add the query and response to conversation memory
    memory.save_context({"query": query}, {"response": response})

    # Display the assistant's response
    st.write("### Assistant's Response")
    st.write(response)
```

```python
import os
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.memory import ConversationBufferMemory
import google.generativeai as genai
from dotenv import load_dotenv

# Load environment variables
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

if not GOOGLE_API_KEY:
    raise ValueError("GOOGLE_API_KEY not found in environment variables")

# Configure Google Generative AI
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize Gemini Pro model
model = genai.GenerativeModel(model_name="gemini-1.0-pro-latest")

# Load the vector store
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
loaded_vectorstore = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)

print(f"Number of documents in the loaded vector store: {loaded_vectorstore._collection.count()}")

# Create a conversation memory
memory = ConversationBufferMemory(k=5)  # Keeps last 5 conversations

# Create a prompt template for query rephrasing
rephrase_template = """Given the following conversation history and the current query, rephrase the query to include relevant context from the conversation:

Conversation history:
{history}

Current query: {query}

Rephrased query:"""

rephrase_prompt = PromptTemplate(
    input_variables=["history", "query"],
    template=rephrase_template
)

# Create a prompt template for the RAG response
rag_template = """You are an AI assistant for a product recommendation system. Use the following pieces of context to answer the human's question. If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context: {context}

Human: {query}
Assistant:"""

# Custom function for generating content with Gemini Pro
def generate_content_with_gemini(prompt):
    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        print(f"Error generating content: {e}")
        return "Sorry, I couldn't generate a response."

# Create an LLMChain for query rephrasing using Gemini Pro
rephrase_chain = LLMChain(llm=generate_content_with_gemini, prompt=rephrase_prompt)

def get_rag_response(query, history):
    # Rephrase the query
    rephrased_query = rephrase_chain.run(history=history, query=query)
    
    # Get relevant documents from the vector store
    results = loaded_vectorstore.similarity_search_with_score(rephrased_query, k=3)
    
    # Handle cases where no documents are returned
    if not results:
        context = ""
    else:
        context = "\n".join([doc.page_content for doc, _ in results])
    
    # Get response from Gemini Pro
    response = generate_content_with_gemini(rag_template.format(context=context, query=rephrased_query))
    
    return response

# Test the framework
conversation_history = ""
while True:
    user_query = input("You: ")
    if user_query.lower() == 'quit':
        break
    
    response = get_rag_response(user_query, conversation_history)
    print("Assistant:", response)
    
    # Update conversation history
    conversation_history += f"Human: {user_query}\nAssistant: {response}\n"
    memory.save_context({"input": user_query}, {"output": response})
```

```python
import os
from dotenv import load_dotenv
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
import google.generativeai as genai

# Load environment variables
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

# Step 1: Configure Google Generative AI
genai.configure(api_key=GOOGLE_API_KEY)

# Step 2: Load the vector store
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
loaded_vectorstore = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)

print(f"Number of documents in the loaded vector store: {loaded_vectorstore._collection.count()}")

# Step 3: Create a conversation memory
memory = ConversationBufferMemory(k=5)  # Keeps last 5 conversations

# Step 4: Create a prompt template for query rephrasing
rephrase_template = """Given the following conversation history and the current query, rephrase the query to include relevant context from the conversation:

Conversation history:
{history}

Current query: {query}

Rephrased query:"""

rephrase_prompt = PromptTemplate(
    input_variables=["history", "query"],
    template=rephrase_template
)

# Step 5: Create a prompt template for the RAG response
rag_template = """You are an AI assistant for a product recommendation system. Use the following pieces of context to answer the human's question. If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context: {context}

Human: {query}
Assistant:"""

rag_prompt = PromptTemplate(
    input_variables=["context", "query"],
    template=rag_template
)

# Step 6: Simulate a query
user_query = "What is the best laptop under 50k?"

# Step 7: Manually rephrase the query using Google Generative AI model
conversation_history = memory.buffer  # Get conversation history (this will be a string)
rephrase_prompt_text = rephrase_prompt.format(history=conversation_history, query=user_query)

rephrase_response = genai.generate_text(model="models/gemini-1.0-pro", prompt=rephrase_prompt_text)
rephrased_query = rephrase_response.generations[0].text
print(f"Rephrased query: {rephrased_query}")

# Update memory with the latest interaction
memory.save_context({"input": user_query}, {"output": rephrased_query})

# Step 8: Search the vector store for relevant documents
docs = loaded_vectorstore.similarity_search(rephrased_query, k=5)
context = "\n".join([doc.page_content for doc in docs])

# Step 9: Generate the final response using Google Generative AI model (RAG approach)
rag_prompt_text = rag_prompt.format(context=context, query=rephrased_query)
rag_response = genai.generate_text(model="models/gemini-1.0-pro", prompt=rag_prompt_text)

final_response = rag_response.generations[0].text
print(f"Final response: {final_response}")

# Step 10: Update memory with final response
memory.save_context({"input": rephrased_query}, {"output": final_response})
```
