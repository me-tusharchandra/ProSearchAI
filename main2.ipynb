```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.schema import Document
import pandas as pd
from tqdm import tqdm

# Load your CSV file
df = pd.read_csv('updated_productData.csv')

# Define the embedding model with the correct configuration
embedding_model_name = "all-MiniLM-L6-v2"
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name)

def create_document(row):
    # Create a document with full content and metadata
    content = f"Product Name: {row['product_name']}" \
              f"Product URL: {row['product_url']}" \
              f"Brand: {row['brand']}" \
              f"Category: {row['product_category_tree']}" \
              f"Retail Price: {row['retail_price']}" \
              f"Discounted Price: {row['discounted_price']}" \
              f"Rating: {row['product_rating']}" \
              f"Overall Rating: {row['overall_rating']}" \
              f"Description: {row['description']}"
    
    metadata = {
        "product_name": row['product_name'],
        "product URL": row['product_url'],
        "brand": row['brand'],
        "category": row['product_category_tree'],
        "retail_price": row['retail_price'],
        "discounted_price": row['discounted_price'],
        "product_rating": row['product_rating'],
        "overall_rating": row['overall_rating']
    }
    
    return Document(page_content=content, metadata=metadata)

# Create a list of Document objects with progress bar
documents = []
for _, row in tqdm(df.iterrows(), total=df.shape[0], desc="Creating Documents"):
    documents.append(create_document(row))

# Create a persistent directory for the Chroma database
persist_directory = "chroma_db"

# Create the Chroma vector store
vector_store = Chroma.from_documents(
    documents=documents,
    embedding=embeddings,
    persist_directory=persist_directory
)

# Persist the vector store
vector_store.persist()

print(f"Vector store saved successfully in '{persist_directory}'")

# For querying
loaded_vectorstore = Chroma(embedding_function=embeddings, persist_directory=persist_directory)
```

```python
import google.generativeai as genai
from langchain.vectorstores import Chroma

# Load the vector store
loaded_vectorstore = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)

# Initialize the Gemini model
model = genai.GenerativeModel(model_name="gemini-1.5-flash")

def generate_response_from_model(user_query, documents, model):
    
    # Build the context using the complete documents
    context = "\n\n".join([
        f"Product Name: {doc.page_content}\n"
        f"Description: {doc.metadata.get('description', 'N/A')}\n"
        f"Brand: {doc.metadata.get('brand', 'N/A')}\n"
        f"Image URL: {doc.metadata.get('imgUrl', 'N/A')}\n"
        f"Product URL: {doc.metadata.get('productURL', 'N/A')}\n"
        f"Retail Price: {doc.metadata.get('retailPrice', 'N/A')}\n"
        f"Discounted Price: {doc.metadata.get('discountedPrice', 'N/A')}\n"
        f"Product Rating: {doc.metadata.get('productRating', 'N/A')}\n"
        f"Overall Rating: {doc.metadata.get('overallRating', 'N/A')}"
        for doc in documents
    ])

    # Construct the prompt
    prompt = f"""
    You are an intelligent shopping assistant with access to a comprehensive product database. Your goal is to help users find, compare, and analyze products based on the following details:

    - **Product Name**: The name of the product.
    - **Description**: A brief description of the product.
    - **Brand**: The brand of the product.
    - **Image URL**: The URL of the product image.
    - **Product URL**: The link to purchase the product.
    - **Retail Price**: The original retail price of the product.
    - **Discounted Price**: The current discounted price of the product.
    - **Product Rating**: The rating of the product.
    - **Overall Rating**: The overall rating of the product.

    User Query: {user_query}

    Here are some related products:
    {context}
    
    Please provide a detailed response to the user's query, including links to products, comparison details if requested, and any other relevant information.
    """

    # Generate the response using the model
    response = model.generate_content(prompt)
    return response

def process_user_query(user_input, vectorstore, model):

    # Step 1: Perform a similarity search in the vector store based on the user input
    search_results = vectorstore.similarity_search(user_input, k=3)  # Retrieve top 5 similar products
    
    if len(search_results) == 0:
        return "No product found. Please try a different query."

    # Step 2: Generate a response from the model based on the user's query and the context
    response = generate_response_from_model(user_input, search_results, model)
    # response = response[1:]
    return response

# Example usage:
user_input = "Can you explain features of Alisha Solid Women's Cycling Shorts."
response = process_user_query(user_input, loaded_vectorstore, model)
```
